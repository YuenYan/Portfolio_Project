```python
# you need to get content through HTML
# HTML --> scraping content from HTML

#import library
import requests
from bs4 import BeautifulSoup as soup
```

```python
# bs4 : enable us to look at markup language
# RSS news feed --> markup language
```

```python
#save url as a string
url = 'https://www.nike.com.hk/man/running/shoe/list.htm?intpromo=PNTP'
url
```

```python
type(requests.get(url).content)
```

```python
# bytes --> sth that python cannot read
nike_html = requests.get(url).content
soup_page = soup(nike_html, 'html.parser')
```

```python
# get the names and prices of our nike
```

```python
#.text return a string
soup_page.title.text
```

```python
soup_page.span.text
# when you call a tag by just simply its name
# you'll onlt get the first instance that this tag is made (only show the first one only)
```

```python
soup_page.findAll('span')
```

```python
# webscrape --> peeling onions
# name & price of our nike shoes
# see where they are stored

# from the element page we see dl tag contains the content we want
# dl tag --> class = "product_list_content"
soup_page.findAll("dl")[0]
# soup_page.findAll("dl") --> return a list of all 'dl' tag
# soup_page.findAll("dl")[0] --> the 1st element of the list
```

```python
shoes_list = soup_page.findAll("dl",{'class':'product_list_content'})
shoes_list
# only get the dl tag with the class as product_list_content
```

```python
# p.s.
soup_page.findAll("dl",{'class':'product_list_content'})[0]
# get the 1st element from the list of the dl tag with the class as product_list_content
```

```python
shoes_list[0]
```

```python
# you should now know: 'findALL' method
```

```python
shoes_list[0].find('dd',{'class':'color666'}).text
# findALL
# try to locate the price
# price --> string --> "HK$799"
```

```python
shoes_list[0].find('a',{'class':'product_list_name'}).find('span', {'class':'up'}).text
# 'a' tag --> hyper links (click the word and you will be direct to the link)
# locate 'a' tag, then from the content of a tag locate the span tag
# .text --> return a string
```

```python
# p.s. for the type
type(shoes_list[0].find('a',{'class':'product_list_name'}).find('span', {'class':'up'}))
# type --> bs4.element.Tag
# if you add .text, if return a string instead
```

```python
# use the for loop to get the content of shoes list (i.e. price & name of all shoes) all at once
for shoe in shoes_list:
  print(shoe.find('dd',{'class':'color666'}).text)
  print(shoe.find('a',{'class':'product_list_name'}).find('span', {'class':'up'}).text)
  print('--'*50) # separate with 100'-'
```

```python
# you can't make it into sheet using print() function (google sheet, excel sheet)

names = []
prices = []

for shoe in shoes_list:
  prices.append(shoe.find('dd',{'class':'color666'}).text)
  names.append(shoe.find('a',{'class':'product_list_name'}).find('span', {'class':'up'}).text)
```

```python
nike_d = {
    'Name':names,
    'Price':prices
}
```

```python
import pandas as pd #panel data
df = pd.DataFrame(nike_d)
df
```

```python
df.info()
# price are string, not ideal (want to be number)
# some have ','
```
